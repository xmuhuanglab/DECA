###

## **Step 2: Setting Folder**

#### **sparse_filepath:** the path you save project's sparse matrix, .mtx format(coo_matrix) is recommanded 
#### **index_path:** the path you save peak names
#### **colnames_path:** the path you save cell barcode
#### **output_path:** the path you'd like to save devided peaks information
#### **patch_num:** the peak number of each patch. 

#### **folder_path**:the path you save your chromosome peak file, It can be created by code above section or without devided
#### **workpath:** the path you like to save all outputs created following
#### **metapath:** the path of metadata including a least barcode and cell-type. Please seperate this file by ',' and name barcode columns as 'Cell', cell-type columns as 'celltype'

```python
sparse_filepath='./data/sparse_matrix.mtx'
index_path='./data/index.csv'
colnames_path='./data/colnames.txt'
output_path='./single-cell_simu/devide_chr_data/devide_chr/'
patch_num=100


folder_path = './single-cell_simu/devide_chr_data/devide_chr/'
workpath='./'
metapath='./data/submeta.csv'
```

###

## **Step 3: Deviding Chromosome**

```python
file_list=devide_chr(sparse_filepath,index_path,colnames_path,output_path,patch_num)
print('Files have been saved into:',end="")
print(file_list)
```

###

## **Step 4: Running Deformer (Training)**

#### **samplenum:** the sample number of simulation. 
#### **patch_size:** the peak number of each patch. 
#### **variance_threshold:** the cut-off of scale variance. 
#### **batch_size:** the batch size of GPU processed. 
#### **epochs:** the iteration number of Deformer. 

```python
Results=Deformer(workpath,metapath,file_list,samplenum=100,patch_size=100,variance_threshold=1,batch_size=1,epochs=50)
```
